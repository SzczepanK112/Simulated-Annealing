```python
kety_coords = (49.88335218571101, 19.22146813090962)


graf_kety = pobierz_graf_osm_z_punktu(
    center_point=kety_coords,
    dist=800,
    dist_type="bbox",
    custom_drogi = ["tertiary", "residential"]
)
graf_kety.rysuj(size_x=15, size_y=15, show_labels=False, show_coords=False, node_size=50, label_font_size=10, show_edge_labels=False, edge_width=2)
print("Graf glowne drogi - liczba krawędzi:", len(graf_kety.krawedzie))
```

```python
snowfall_forecast = [4, 8, 5, 3]

m1 = Machine(speed=25) # W km/h
m2 = Machine(speed=15)

machines = [m1, m2]

Tmax = 1 # W godzinach

problem = RoadClearingProblem(snowfall_forecast, graf_kety, machines, Tmax)

initial_temperature = 30
cooling_rate = 0.95
max_iterations = 1000

best_solution, best_danger, diagnostics = problem.simulated_annealing_2(initial_temperature, cooling_rate, max_iterations)
m1_best = best_solution[0].route
m2_best = best_solution[1].route
print("\n")
print("Najlepsze rozwiązanie:")
print(m1_best)
print(m2_best)
print("Poziom zagrożenia:", best_danger)
```
